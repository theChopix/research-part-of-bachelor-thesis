<h2 align="center"> <font color='gray'>catboost Built-in Visualization</font></h2>


<div class="alert alert-block alert-success">  
<b>Description:</b> CatBoost library itself provides one built-in function to visualize decision trees. 
</div>

```python
import json
from catboost import CatBoostRegressor
```

```python
model = CatBoostRegressor();
model.load_model("../../data/model.json", format='json');
```

<h3 align="center"> <font color='gray'>catboost Visualization Function plot_tree()</font></h3>

```python
viz = model.plot_tree(tree_idx=0)
viz
```

<div class="alert alert-block alert-success">  
<b>Description:</b> General decision tree visualization with indices of split features together with split values in its internal nodes and values in leaves.
</div>

```python
viz.render(filename="visualizations/catboost_viz");
```

<div class="alert alert-block alert-success">
